# Polinomios de Lagrange
El polinomio interpolante de lagrange
$$P(x)=\sum_{i=0}^{n}L_i(x)f(x_i)=\sum_{i=0}^{n}L_i(x)y_i$$

y Los polinomios de lagrange

$$L(x)=\prod_{j=0, j\neq i}^{n}\frac{(x-x_j)}{(x_i-x_j)}$$

```python
import sympy as sp
x = sp.symbols('x')

def lagrange(xdata, ydata):
    P = 0
    n = len(xdata)
    for i in range(n):
        Li = 1
        for j in range(n):
            if j!= i:
                Li *= ( x - xdata[j])/(xdata[i] - xdata[j]) 
        P += Li * ydata[i]
    return sp.expand(P) 

P = sp.lambdify(x, lagrange(x_data, y_data))
ux = np.linspace(min(x_data), max(x_data), 100)
plt.plot(x_data, y_data, 'gp', label = 'Datos observados')
plt.plot(ux, P(ux), 'r', label = 'Polinomio de Lagrange')
plt.legend()
```
