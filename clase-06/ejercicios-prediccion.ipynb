```python
import matplotlib.pyplot as plt
import numpy as np


x = [1]
h = 0.01
k = 0.0009906
t_inicial= 0
t_final = t_inicial+h

while t_final < 6:
    x.append(h*k*x[-1]*(1000-x[-1])+x[-1])
    t_final += h

plt.subplot(121)
plt.plot(np.arange(t_inicial, 6+h, h), x)
plt.subplot(122)
plt.plot(np.arange(t_inicial, 6+h, h), 1000-np.array(x))
plt.grid()


```
