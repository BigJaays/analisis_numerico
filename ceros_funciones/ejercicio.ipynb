```python
from ceros import biseccion

import numpy as np
import matplotlib.pyplot as plt

x = 1.7
g = 32.17
t = 1

funcion = lambda w: - g / (2*w**2) * ((np.exp(w*t) -np.exp(-w*t)) / 2 - np.sin(w*t)) - x

w = np.linspace(-3, -1e-3, 100)
plt.plot(w, funcion(w), 'b')
plt.axhline(0, color = 'red', linewidth = 1, linestyle = 'dashed')
plt.grid()

valor = biseccion(funcion, -1, -1e-3, 1e-5)
print(valor)
```

```python
"""
La suma de dos numeros es 20. Si a cada uno se le suma la raiz cuadrada, el producto de las dos sumas es de 155.55.
Determine los dos numeros con una exactitud de 10^-4.
"""

import numpy as np
import matplotlib.pyplot as plt

funcion = lambda x: (x+np.sqrt(x))*(20 - x + np.sqrt(20-x)) - 155.55

x = np.linspace(0, 20, 100)
plt.plot(x, funcion(x), 'b')
plt.axhline(0, color = 'red', linewidth = 1, linestyle = 'dashed')
plt.ylim(-25, 25)
plt.grid()

iteraciones_x, x = biseccion(funcion, 5, 7.5, 1e-4)
print(f" Las iteraciones fueron: {iteraciones_x} y x es: {x}")

iteraciones_y, y = biseccion(funcion, 12.5, 15, 1e-4)
print(f" Las iteraciones fueron: {iteraciones_y} y y es: {y}")

print(f"Los dos numeros son: {x} y {y} y su suma da {x+y}")


```
